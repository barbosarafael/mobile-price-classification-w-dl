# 0) Setup

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
pd.set_option('display.max_columns', None)
```

# 1) Carregando os dados

```python
train = pd.read_csv('../data/train.csv')

display(train.head())

display(train.info())
```

## 2) Análises

### 2.1) Univariada

```python
#---- Variáveis "categóricas"

# São flags apesarem de estar em formato inteiro

str_cols = ['blue', 'dual_sim', 'four_g', 'm_dep', 'three_g', 'touch_screen', 'wifi']

for col in str_cols:
    
    print(f'>>>>>>>>>> Variável: {col}')
    
    df_temp = train[col]\
        .value_counts()\
        .reset_index()
        
    df_temp_perc = train[col]\
        .value_counts(normalize = True)\
        .reset_index()\
        .assign(\
                proportion = lambda x: x['proportion'] * 100
            )
        
    df_display = df_temp\
        .merge(df_temp_perc, on = col, how = 'left')
        
    display(df_display)
```

```python
# Variáveis numéricas

numbers_cols = train.head(1).drop(columns = ['price_range'] + str_cols).columns.tolist()
```

```python
train[numbers_cols].describe()
```

```python
for col in numbers_cols:
    
    print(f'>>>>>>>>>> Variável: {col}')
    
    fig = px.histogram(train, x = col, title = col)
    fig.show()
```

### 2.2) Multivariada -> Relacionamento com a variável `price_range` (target)

- Resumo: no geral as variáveis possuem relacionamento iguais entre suas classes e a variável target

```python
# Variáveis categóricas

for col in str_cols:
    
    df_temp = train[[col, 'price_range']]\
        .value_counts()\
        .reset_index()\
        .assign(\
                price_range = lambda x: x['price_range'].astype(str)
            )
        
    df_temp[col] = df_temp[col].astype(str)

    fig = px.bar(data_frame = df_temp, x = 'price_range', y = 'count', color = col)
    fig.show()
```

```python
# Variáveis numéricas

for col in numbers_cols:
    
    fig = px.box(train,
                x = col,
                color = 'price_range',
                category_orders = {'price_range': [0, 1, 2, 3]})
    
    fig.show()
```
